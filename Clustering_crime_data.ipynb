(b) K - MEANS


STEP 1 = Data Pre-processing :

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
crime_cases = pd.read_csv('/content/crime_data.csv')
crime_cases.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler_2 = StandardScaler()
x_2 = std_scaler_2.fit_transform(crime_cases.iloc[:,1:])
print(x_2)
```

STEP 2 = Finding the optimal number of clusters using the Elbow Method :

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i,random_state = 0)
    kmeans.fit(x_2)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1, 11), wcss)
plt.scatter(range(1, 11), wcss,color = 'red')
plt.title('ELBOW METHOD')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

STEP 3 = Training the K - means clustering model :

```python
km = KMeans(4, random_state = 60)
km.fit(x_1)
```

```python
km.labels_
```

```python
crime_cases['clusters_2'] = km.labels_
km.cluster_centers_
```

```python
crime_cases.groupby('clusters_2').agg(['mean']).reset_index()
```

STEP 4 = Visualizing the clusters :

```python
crime_cases[crime_cases['clusters_2']==0]
```

```python
crime_cases[crime_cases['clusters_2']==1]
```

```python
crime_cases[crime_cases['clusters_2']==2]
```

(c) DBSCAN


STEP 1 = Data Pre-processing :

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
crime_cases = pd.read_csv('/content/crime_data.csv')
crime_cases.head()
```

```python
df = pd.DataFrame(crime_cases)
df.head()
```

```python
df.columns
```

```python
df1 = df.drop(['Unnamed: 0'],axis = 1)
df1.head()
```

STEP 2 = Finding the optimal number of clusters using the DBSCAN :

```python
sns.histplot(crime_cases.Assault)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler_3 = StandardScaler().fit(df1)
x_3 = std_scaler_3.transform(df1)
x_3
```

STEP 3 = Training the DBSCAN clustering model :

```python
db = DBSCAN(eps = 1.1,min_samples = 3)
db.fit(x_3)
```

```python
db.labels_
```

```python
df2 = pd.DataFrame(db.labels_,columns = ['clusters_3'])
df2.head()
```

```python
crime_cases['clusters_3'] = df2['clusters_3']
crime_cases.head()
```

```python
df2.clusters_3.value_counts()
```

```python
crime_cases.groupby(['clusters_3']).agg(['mean']).reset_index()
```
